```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
dados = response.json()
```

```python
df = pd.DataFrame(dados)
df_filter = df[["id", "name", "username", "email"]]
df_filter
```

```python
df_filter.to_excel("user_name_rpa.xlsx", index=False)
plt.figure(figsize=(10, 5))
sns.barplot(x="name", y="id", data=df_filter, hue="name", palette="mako", legend=False)
plt.xticks(rotation=45)
plt.show()
```
